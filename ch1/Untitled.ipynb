```python
HaccpInfor = {}

namelist = []
companylist=[]
businessitemnmlist=[]
productgblist=[]
prdlstreportnolist=[]


page = 1

#haccp data 잔뜩

for page in range(1200,1220):
    url = 'http://apis.data.go.kr/B553748/CertPrdListService/getCertPrdListService?ServiceKey=3e%2BxITG3ESne4UwH7%2FxQ%2BNajrByAAxIpr9HmQSk5M%2Fte5wTA5OWPXLonke13WXBRhiZatSAtdfhSGu9dvnD8YA%3D%3D&numOfRows=100&pageNo='+str(page)
    req = requests.get(url)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    
    productGb=soup.find_all('productgb')
    prdlstNm = soup.find_all('prdlstnm')
    businessitemNm = soup.find_all('businessitemnm')
    #imgurl=soup.find_all('imgurl1')
    company = soup.find_all('company')
    prdlstReportNo = soup.find_all('prdlstreportno')
    
    for code in productGb:
        productgblist.append(code.text)
    for code in prdlstNm:
        namelist.append(code.text)
    for code in businessitemNm:
        businessitemnmlist.append(code.text)
  #  for code in imgurl:
       # imgurllist.append(code.text)
    for code in company:
        companylist.append(code.text)
    for code in prdlstReportNo:
        prdlstreportnolist.append(code.text)

    

HaccpInfor['PrdlstNm'] = namelist
HaccpInfor['BusinessitemNm']= businessitemnmlist
HaccpInfor['ProductGb'] = productgblist
HaccpInfor['Company'] = companylist
HaccpInfor['prdlstreportno'] = prdlstreportnolist

df = pandas.DataFrame(HaccpInfor)

df.to_excel("haccpdata.xlsx",  encoding='utf-8')
```

```python
HaccpInfor2 = {}
rawmtrllist=[]
imgurllist=[]
prdlstreportnolist=[]

for page in range(10,30):
    url = 'http://apis.data.go.kr/B553748/CertImgListService/getCertImgListService?ServiceKey=3e%2BxITG3ESne4UwH7%2FxQ%2BNajrByAAxIpr9HmQSk5M%2Fte5wTA5OWPXLonke13WXBRhiZatSAtdfhSGu9dvnD8YA%3D%3D&numOfRows=100&pageNo='+str(page)
    req = requests.get(url)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    




    imgurl=soup.find_all('imgurl1')
    prdlstReportNo = soup.find_all('prdlstreportno')
    
    rawmtrl=soup.find_all('rawmtrl')
    for code in rawmtrl:
        rawmtrllist.append(code.text)
    for code in imgurl:
        imgurllist.append(code.text)
    for code in prdlstReportNo:
        prdlstreportnolist.append(code.text)

    
HaccpInfor2['prdlstreportno'] = prdlstreportnolist
#HaccpInfor2['imgurll'] = imgurllist
HaccpInfor2['rawmtrl'] = rawmtrllist
df2 = pandas.DataFrame(HaccpInfor2)

df2.to_excel("haccpdataimage.xlsx",  encoding='utf-8')
```
